## 1. Energy Burden vs Protocol-Offset Simulator (Matching Figure 9.2)
This section allows interactive exploration of the "Energy Burden vs Protocol-Offset" curve, directly reproducing Figure 9.2 from the book. The Protocol Offset is a key mechanism to reduce energy burden.

```python
def plot_energy_burden(income_level_mwh_per_euro, protocol_offset_rate_slider):
    protocol_offsets = np.linspace(0, core.protocol_offset_max, 100)
    energy_burdens = [core.calculate_energy_burden(income_level_mwh_per_euro, po) for po in protocol_offsets]

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(protocol_offsets * 100, np.array(energy_burdens) * 100, label="Energy Burden", color='blue')
    
    # Highlight current slider value
    current_eb = core.calculate_energy_burden(income_level_mwh_per_euro, protocol_offset_rate_slider)
    ax.plot(protocol_offset_rate_slider * 100, current_eb * 100, 'ro', markersize=8, label=f'Current: {current_eb*100:.2f}%')

    ax.set_xlabel("Protocol Offset (%)", fontsize=12)
    ax.set_ylabel("Energy Burden (%)", fontsize=12)
    ax.set_title("Figure 9.2: Energy Burden vs Protocol-Offset", fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()
    plt.ylim(0, 25) # Assuming max burden is around 20-25% for visualization
    plt.xlim(0, core.protocol_offset_max * 100)
    plt.show()

    display(Markdown(f"**Current Simulation Result:** With **{protocol_offset_rate_slider*100:.0f}%** Protocol Offset and **{income_level_mwh_per_euro:.2f} MWh/€** Energy Intensity, the Energy Burden is: **{current_eb*100:.2f}%**"))

interact(plot_energy_burden,
         income_level_mwh_per_euro=FloatSlider(min=0.01, max=0.10, step=0.01, value=0.05, description='Energy Intensity (MWh/€)', continuous_update=False),
         protocol_offset_rate_slider=FloatSlider(min=0.0, max=core.protocol_offset_max, step=0.01, value=0.3, description='Protocol Offset Rate', continuous_update=False))
```

## 2. Utility Equation Calculator
This section allows you to calculate the Protocol-Enabled Utility, demonstrating how it aims to reverse the "Demographic Drain" as described in the book.

```python
def calculate_utility(demographic_drain_rate_slider, aep_node_efficiency_slider):
    utility_gain = core.protocol_enabled_utility_equation(demographic_drain_rate_slider, aep_node_efficiency_slider)
    display(Markdown(f"**Calculated Protocol-Enabled Utility Gain:** **{utility_gain:.2f}** (higher is better)

*This metric quantifies how effectively the protocol mitigates demographic drain.*"))

interact(calculate_utility,
         demographic_drain_rate_slider=FloatSlider(min=0.0, max=0.1, step=0.005, value=core.demographic_drain_rate, description='Demographic Drain Rate', continuous_update=False),
         aep_node_efficiency_slider=FloatSlider(min=0.0, max=1.0, step=0.01, value=core.aep_node_efficiency, description='AEP Node Efficiency', continuous_update=False))
```

## 3. Polynomial of Social Unrest
This section explores the "Polynomial of Social Unrest," illustrating the book's concept of how energy burden can lead to societal instability.

```python
def calculate_social_unrest(unrest_factor_slider, energy_burden_for_unrest_slider):
    social_unrest = core.polynomial_of_social_unrest(unrest_factor_slider, energy_burden_for_unrest_slider)
    display(Markdown(f"**Calculated Social Unrest:** **{social_unrest:.4f}**

*This value represents the potential for social unrest based on the given energy burden and unrest factor.*"))

interact(calculate_social_unrest,
         unrest_factor_slider=FloatSlider(min=0.1, max=1.0, step=0=0.05, value=core.unrest_factor, description='Unrest Factor', continuous_update=False),
         energy_burden_for_unrest_slider=FloatSlider(min=0.0, max=0.5, step=0.01, value=0.15, description='Energy Burden (for Unrest)', continuous_update=False))
```
